### A university wants to automate their admission process. Students are admitted based on marks scored in a qualifying exam.

A student is identified by student id, age and marks in qualifying exam. Data are valid, if:

  a. Age is greater than 20

  b. Marks is between 0 and 100 (both inclusive)

A student qualifies for admission, if

  a. Age and marks are valid and

  b. Marks is 65 or more

Write a python program to represent the students seeking admission in the university.

Business rules

```python
class Student:
    def __init__(self):
        self.__student_id = None
        self.__marks = None
        self.__age = None
        
    def Validate_marks(self):
        if self.__marks:
            if self.__marks >=0 and self.__marks<=100:
                return True

        
    def Validate_age(self):
        if self.__age >20:
            return True
        
    def Check_qualification(self):
        if self.Validate_age() and self.Validate_marks():
            if self.__marks >= 65:
                print("Eligible")
                return True
            else:
                print("Not eligible")
        else:
            print("Not eligible")
            
    def set_student_id(self,s_id):
        self.__student_id = s_id
        
    def set_marks(self,mark):
        if type(mark) == int and mark>=0:
            self.__marks = mark
        else:
            print("Mark should be in Int format and greater than or euqal to 0")
            
    def set_age(self,age):
        if type(age) == int:
            self.__age= age
        else:
            print("Age should be in Int format")
            
    def get_marks(self):
        return self.__marks
    
    def get_age(self):
        return self.__age
    
    def get_student_id(self):
        return self.__student_id
```

```python
Naveen.set_marks(80)
```

```python
Naveen.get_marks()
```

```python
Naveen.Check_qualification()
```

### Continuing with the previous scenario, a student eligible for admission has to choose a course and pay the fees for it. If they have scored more than 85 marks in qualifying exam, they get 25% discount on fees.

Valid course ids and fees are given below:

Extend the program written in the previous assignment to include the above requirement.

```python
class Student:
    def __init__(self):
        self.__student_id = None
        self.__marks = None
        self.__age = None
        self.__course_id = None
        self.__student_fee = None
        self.__fees = {"1001":25575,"1002":15500}
        
    def choose_course(self,course_id):
        if course_id in self.__fees:
            self.set_course_id(course_id)
            if self.__marks > 85:
                self.__student_fee = self.__fees[course_id] - (self.__fees[course_id]*(25/100))
            return True
        else:
            print("Not in list")
            return False
            
    def set_course_id(self,course_id):
        self.__course_id = course_id
    
    def Validate_marks(self):
        if self.__marks:
            if self.__marks >=0 and self.__marks<=100:
                return True

        
    def Validate_age(self):
        if self.__age >20:
            return True
        
    def Check_qualification(self):
        if self.Validate_age() and self.Validate_marks():
            if self.__marks >= 65:
                print("Eligible")
                return True
            else:
                print("Not eligible")
        else:
            print("Not eligible")
            
    def set_student_id(self,s_id):
        self.__student_id = s_id
        
    def set_marks(self,mark):
        if type(mark) == int and mark>=0:
            self.__marks = mark
        else:
            print("Mark should be in Int format and greater than or euqal to 0")
            
    def set_age(self,age):
        if type(age) == int:
            self.__age= age
        else:
            print("Age should be in Int format")
            
    def get_marks(self):
        return self.__marks
    
    def get_age(self):
        return self.__age
    
    def get_student_id(self):
        return self.__student_id
    
    def get_course_id(self):
        return self.__course_id
    
    def get_student_fee(self):
        return self.__student_fee
```
